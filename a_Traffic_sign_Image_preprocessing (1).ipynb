```python
## packages
import os as Trfic_P1 ##helps with file & directory operations
from PIL import Image as Trfic_P2 ##helps with image processing
import matplotlib.pyplot as Trfic_P3 ##helps to plot and visualize images
```

Folder-1

```python
#  PPM images path
Trfic_PPM_fls = "/content/drive/MyDrive/Final Project/Code-20250403T153521Z-001/Code/Images/00001"

Trfic_JPG_fls = "/content/drive/MyDrive/Final Project/Code-20250403T153521Z-001/Code/Traffic_JPG_images/1"

if not Trfic_P1.path.exists(Trfic_JPG_fls):
    Trfic_P1.makedirs(Trfic_JPG_fls)

# Loop through all the files in the folder
for PPM in Trfic_P1.listdir(Trfic_PPM_fls):

    if PPM.lower().endswith(".ppm"):
        ppm_path = Trfic_P1.path.join(Trfic_PPM_fls, PPM) ## builds the full path to .ppm file
        with Trfic_P2.open(ppm_path) as open_ppm:

            # defining JPG extension
            jpg = Trfic_P1.path.splitext(PPM)[0] + ".jpg"
            jpg_path = Trfic_P1.path.join(Trfic_JPG_fls, jpg)
            open_ppm.convert("RGB").save(jpg_path, "JPEG")
```

```python
Trfic_JPG_fls_1 = "/content/drive/MyDrive/Final Project/Code-20250403T153521Z-001/Code/Traffic_JPG_images/1"

Trfic_JPG = [f for f in Trfic_P1.listdir(Trfic_JPG_fls_1) if f.endswith(('jpg'))]
Trfic_JPG = Trfic_JPG[:10]

# display images
fi_Trfic, ax_Trfic= Trfic_P3.subplots(2, 5, figsize=(15, 6))
ax_Trfic = ax_Trfic.flatten()

for jj, jpg in enumerate(Trfic_JPG): ##iterates through a list of JPG files and displays them using matplotlib.
    PATH = Trfic_P1.path.join(Trfic_JPG_fls_1, jpg)##builds the full path for each image using os.path.join
    sign = Trfic_P2.open(PATH) ##opens the image using PIL

    # Display the image
    ax_Trfic[jj].imshow(sign)
    ax_Trfic[jj].axis('off')
    ax_Trfic[jj].set_title(jpg)

Trfic_P3.tight_layout()
Trfic_P3.show()
```
